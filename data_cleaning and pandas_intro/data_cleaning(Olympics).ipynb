```python
import pandas as pd
import numpy as np
```

```python
bios=pd.read_csv("https://raw.githubusercontent.com/KeithGalli/Olympics-Dataset/master/athletes/bios.csv")
```

```python
bios.head()
```

```python
df=bios.copy()
```

```python
df.head()
```

```python
df['name']=df['Used name'].str.replace('•',' ')
```

```python
df[['Height_cm','Weight_kg']]=df['Measurements'].str.split('/',expand=True)
df['Weight_kg'].fillna(df['Height_cm'],inplace=True)
df['Height_cm']=pd.to_numeric(df['Height_cm'].str.strip(' cm'),errors='coerce')
df['Weight_kg']=pd.to_numeric(df['Weight_kg'].str.strip(' kg'),errors='coerce')
```

```python
df.tail()
```

```python
df[['name','Height_cm']].sort_values('Height_cm',ascending=False).iloc[0:6].reset_index(drop=True)
```

```python
date_pattern=r'(\d+ \w+ \d{4}|\d{4}|\w+ \d{4})'
df['born_dates']=df['Born'].str.extract(date_pattern)
df['born_dates']=pd.to_datetime(df['born_dates'],errors='coerce')

df['died_date']=df['Died'].str.extract(date_pattern)
df['died_date']=pd.to_datetime(df['died_date'],errors='coerce')
```

```python
df.head()
```

```python
location_pattern = r"in\s+([\w\s\-']+),\s+([\w\s\-é]+)\s+\(([A-Z]{3})\)"
```

```python
df[['born_city','born_region','born_country']]=df['Born'].str.extract(location_pattern,expand=True)
```
